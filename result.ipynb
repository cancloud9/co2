**讀取資料**

```python
import pandas as pd

df = pd.read_csv('sensorData.csv')
data = list()
target = [0] * 34038
for i in range(34038):
    data.append(list(df.iloc[i,1:]))

target.extend([1] * 30763)
for i in range(34037, 64800):
    data.append(list(df.iloc[i,1:]))
    
print(len(df), 'Datas')
```

**繪製資料散點圖**

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(x = [d[0] for d in data], y = [d[1] for d in data], c = target, marker = 'o') 
plt.xlabel('light') 
plt.ylabel('co2')
plt.show()
```

**創建k-nn分類器 並輸出分類結果**

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=13)
model.fit(data, target)
predict = model.predict(data)

err = 0
for i in range(len(df)):
    if predict[i] != target[i]:
        err += 1

print('accuracy:', 1 - (err / (len(df))))
```
